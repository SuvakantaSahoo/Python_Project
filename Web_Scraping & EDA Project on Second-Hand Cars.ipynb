# EXTRACTING DATA FROM THE WEBSITE

```python
# get the required text for all used cars

for x in soup.find_all('div', attrs={'class':" _1l4fi"}):
    if x is None:
        print('Data not Found')
    else:
        print(x.text, '\n\n')
```

```python
# get the price for all Cars

for x in soup.find_all('div', attrs={'class':"_1l4fi"}):
    price = x.find('div', attrs={'class': "_7udZZ"})
    if price is None:
        print('Data not Found')
    else:
        print(price)
        print(price.text, '\n\n')
```

```python
# get the title for all Cars(Year & Car Company)

for x in soup.find_all('div', attrs={'class':"_1l4fi"}):
    title = x.find('h2', attrs={'class': "_3FpCg"})
    if title is None:
        print('Data not Found')
    else:
        print(title)
        print(title.text, '\n\n')
```

```python
# get the version & transmission for all cars

for x in soup.find_all('div', attrs={'class':"_1l4fi"}):
    ver= x.find('p', attrs={'class': "cvakB"})
    if ver is None:
        print('Data not Found')
    else:
        print(ver.text)
    
# get the KM,OWNER,FUEL for all cars

for x in soup.find_all('div', attrs={'class':"_1l4fi"}):
    km = x.find('ul', attrs={'class': "bVR0c"})
    if km is None:
        print('Data not Found')
    else:
        print(km.text)
```

```python
for x in soup.find_all('div', attrs={'class':"_1l4fi"}):
    emi = x.find('div', attrs={'class': "_2HFRN"})
    if emi is None:
        print('Data not Found')
    else:
        print(emi.text, end= '\n\n')

```

```python
titles=[] #List to store name of the product
prices=[] #List to store price of the product
version=[] #List to store rating of the product
kms = [] #List to store the number of ratings & reviews of the product
EMI = [] # to store the specifications
pgno = []


for i in range(1,34):
    start_time = time.time()
    URL = 'https://www.cars24.com/buy-used-cars-new-delhi/'
    pagecontent = requests.get(URL).text
    soup = BeautifulSoup(pagecontent)
    
    for x in soup.find_all('div', attrs={'class':"_1l4fi"}):
        
        title = x.find('div', attrs={'class': "_15rKe"})
        if title is None:
            titles.append(np.NaN)
        else:
            titles.append(title.text)
            
        price = x.find('div', attrs={'class': "_7udZZ"})
        if price is None:
            prices.append(np.NaN)
        else:
            prices.append(price.text)
        
         
        ver = x.find('p', attrs={'class': "cvakB"})
        if ver is None:
            version.append(np.NaN)
        else:
            version.append(ver.text)
        
            
        km = x.find('ul', attrs={'class': "bVR0c"})
        if km is None:
            kms.append(np.NaN)
        else:
            kms.append(km.text)
            
        emi = x.find("div", attrs={'class': "_2HFRN"})
        if emi is None:
            EMI.append(np.NaN)
        else:
            EMI.append(emi.text)
            
        pgno.append(i)
        
    print('Page {} completed in {} seconds'.format(i, time.time()-start_time))
```

# CLEANING THE DATA

```python
df = pd.DataFrame({'Company': titles, 'Brand' : '', 'Model': '', 'Version': version, 'Transmission': '' , 'Fuel': '',  'KM': '' , 'Year': '' , 'Owner': '', 'Price': prices , 'KMS': kms, 'EMI': EMI})
```

```python
# Extract Kilometers
pattern = r'[\s\S]+km'
df['KM']=df.KMS.apply(lambda x:re.compile(pattern).findall(x))
```

```python
# Extract Year
pattern = r'[0-9]{4}'
df['Year']=df.Company.apply(lambda x:re.compile(pattern).findall(x))
```

```python
# Filter year from company
pattern = r'[A-Z][a-z ]+'
df['Company']=df.Company.apply(lambda x:re.compile(pattern).findall(x))
df['Company']=df.Company.apply(lambda x:' '.join(x))
```

```python
# Extract Brand from Company
pattern = r'^\w+'
df['Brand']=df.Company.apply(lambda x:re.compile(pattern).findall(x))
```

```python
pattern = r'[0-9][a-z/s]{2}'
df['Owner']=df.KMS.apply(lambda x:re.compile(pattern).findall(x))
```

```python
# Extract Fuel Type from KMS
df['Fuel'] = df.KMS.apply(lambda x: x[-6:])
```

```python
# Extract Transmission Type from version
pattern = r'\w+$'                                                            
df['Transmission']=df.Version.apply(lambda x:re.compile(pattern).findall(x))
```

```python
# Extract Version
pattern = r'[A-Z]{2,}'
df['Version'] = df.Version.apply(lambda x:re.compile(pattern).findall(x))
```

```python
# Extract Modal from Company
df['Model'] = df.Company.apply(lambda x:x.split()[1:-1])
df['Model'] = df.Model.apply(lambda x:' '.join(f'{w}' for w in x))
```

```python
# Clean Rows which are not used anymore
del df['Company']
del df['KMS']
```

```python
# Clean and format columns
df['Brand']=df['Brand'].str[0]
df['KM']=df['KM'].str[0].str.replace('km' , '').str.replace(',', '').astype(int)
df['Price']=df['Price'].str.replace('₹' , '').str.replace(',', '').astype(int)
df['EMI']=df['EMI'].str.replace('₹' , '').str.replace(',', '').str.replace('/month' , '').astype(int)
df['Version']=df['Version'].apply(lambda x:', '.join(x))
df['Year']=df['Year'].str[0].astype(int)
df['Owner']=df['Owner'].str[0]
df['Transmission']=df['Transmission'].str[0]

df = df[df['Brand'] != '']
df = df.loc[df['Model'] != '']
df = df.loc[df['Version'] != '']
df = df.loc[df['Transmission'] != '']
df = df.loc[df['Fuel'] != '']
df = df.loc[df['KM'] != '']
df = df.loc[df['Year'] != '']
df = df.loc[df['Owner'] != '']
df = df.loc[df['Price'] != '']
df = df.loc[df['EMI'] != '']
df = df[df[['Brand', 'Model', 'Version' , 'Transmission' , 'Fuel' , 'KM' , 'Year' , 'Owner' , 'Price' , 'EMI']].notnull().all(1)]
```

# Handling Outliers

Outliers are those data points which differs significantly from other observations present in given dataset. It can occur because of variability in measurement and due to misinterpretation in filling data points.

How to detect Outliers?
##### 1.IQR Method

 In this method by using Inter Quartile Range(IQR), we detect outliers. IQR tells us the variation in the data set. Any       value, which is beyond the range of -1.5 x IQR to 1.5 x IQR treated as outliers.

- Q1 represents the 1st quartile/25th percentile of the data.
- Q2 represents the 2nd quartile/median/50th percentile of the data.
- Q3 represents the 3rd quartile/75th percentile of the data.
- (Q1–1.5IQR) represent the smallest value in the data set and (Q3+1.5IQR) represent the largest value in the data set
